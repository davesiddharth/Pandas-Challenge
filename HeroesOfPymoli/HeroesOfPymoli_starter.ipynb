```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

```python
purchase_data.head()
```

```python
total_player = len(purchase_data["SN"].unique())
total_player
```

## Player Count

* Display the total number of players


```python
total_players = pd.DataFrame({'Total Players':[total_player]})
total_players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_item = purchase_data["Item ID"].value_counts()
number_of_unique_item = unique_item.count()
#number_of_unique_item

average_price = purchase_data["Price"].mean()
#average_price

number_of_purchases = purchase_data["Item ID"].count()
#number_of_purchases

total_revenue = purchase_data["Price"].sum()
#total_revenue

purchase_analysis = pd.DataFrame({'Number of Unique Items': number_of_unique_item,
                                  'Average Price':average_price,
                                  'Number of Purchases':number_of_purchases,
                                  'Total Revenue':total_revenue
                                 },index=[0])
purchase_analysis.style.format({'Average Price':'${:,.2f}',
                               'Total Revenue':'${:,.2f}'})
```

## Gender Demographics

```python
gender_groupby = purchase_data.groupby(["Gender"])
total_count = gender_groupby.nunique()["SN"]
#total_count

percentage_of_player = total_count/total_player *100
#percentage_of_player

gender_demographics = pd.DataFrame({'Total Count':total_count,
                                    'Percentage of Players':percentage_of_player})

formatted_gender = gender_demographics.sort_values(["Total Count"], ascending=False)

formatted_gender.style.format({'Percentage of Players':'{:,.2f}%'})
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins =[0,9.9,14.99,19.99,24.99,29.99,34.99,39.99,100]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29","30-34","35-39","40+"]

purchase_data["Age Ranges"] = pd.cut(purchase_data["Age"], bins, labels=group_names)
#purchase_data

groupby_age_count = purchase_data.groupby(['Age Ranges'])
age_total_count = groupby_age_count["SN"].nunique()

age_percentage_of_player = age_total_count/total_player * 100
#age_percentage_of_player
```

```python
age_demo = pd.DataFrame({
                        "Total Count":age_total_count,
                        "Percentage of Players":age_percentage_of_player,
                        })
age_demo.index.name = None
age_demo.style.format({'Percentage of Players':'{:,.2f}%'})
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
bins =[0,9.9,14.99,19.99,24.99,29.99,34.99,39.99,100]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29","30-34","35-39","40+"]

purchase_data["Age Ranges"] = pd.cut(purchase_data["Age"], bins, labels=group_names)
#purchase_data

groupby_age_count = purchase_data.groupby(['Age Ranges'])
purchase_count = groupby_age_count["SN"].count()
#purchase_count

avg_price = groupby_age_count['Price'].mean()
#avg_price

total_price = groupby_age_count['Price'].sum()
#total_price

avg_purchase_price_person = total_price/age_total_count
avg_purchase_price_person
```

```python
age_purchase = pd.DataFrame({
                            "Purchase Count":purchase_count,
                            "Average Purchase Price":avg_price,
                            "Total Purchase Value":total_price,
                            "Avg Total Purchase per Person":avg_purchase_price_person,
                            })
age_purchase.style.format({
                            "Average Purchase Price":'${:,.2f}',
                            "Total Purchase Value":'${:,.2f}',
                            "Avg Total Purchase per Person":'${:,.2f}'
                            })
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
sn_group = purchase_data.groupby(['SN'])
sn_purchase_count = sn_group['Purchase ID'].count()
#sn_purchase_count

sn_avg_price = sn_group['Price'].mean()
#sn_avg_price

sn_total_purchase = sn_group['Price'].sum()
sn_total_purchase.head()
```

```python
top_purchase = pd.DataFrame({
                            "Purchase Count":sn_purchase_count,
                            "Average Purchase Price":sn_avg_price,
                            "Total Purchase Value":sn_total_purchase,
                            })
formatted_top_purchase = top_purchase.sort_values(["Total Purchase Value"], ascending=False).head()
formatted_top_purchase.style.format({
                                    "Average Purchase Price":'${:,.2f}',
                                    "Total Purchase Value":'${:,.2f}',
                                    })
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popular_item=purchase_data[['Item ID','Item Name','Price']]
popular_item.head()

popular_group = popular_item.groupby(['Item ID','Item Name'])
popular_group

popular_purchase_count = popular_group['Price'].count()
popular_purchase_count

popular_total_purchase = popular_group['Price'].sum()
popular_total_purchase

popular_item_price = popular_total_purchase/popular_purchase_count
popular_item_price
```

```python
popular_purchase = pd.DataFrame({
                            "Purchase Count":popular_purchase_count,
                            "Item Price":popular_item_price,
                            "Total Purchase Value":popular_total_purchase
                            })
formatted_popular_purchase = popular_purchase.sort_values(["Purchase Count"], ascending=False).head()
formatted_popular_purchase.style.format({
                                    "Item Price":'${:,.2f}',
                                    "Total Purchase Value":'${:,.2f}',
                                    })
```
